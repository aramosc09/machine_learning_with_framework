```python
season_data = nfl.import_seasonal_data([2023])
season_data
```

```python
season_data_merge = pd.merge(season_data, players_data[['player_id','season','season_type','player_display_name','recent_team','position_group','headshot_url']].drop_duplicates(subset=['player_id','season','season_type']), on=['player_id','season','season_type'], how='left')
season_data_merge
```

```python
season_data_merge['position_group'].value_counts()
```

```python
filtered_positions = season_data_merge[season_data_merge['position_group'].isin(['QB', 'WR','RB','TE'])]
filtered_positions
```
